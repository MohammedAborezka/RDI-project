```python
# Load The Data
import pandas as pd
import csv
from ast import literal_eval
df = pd.read_csv("/content/drive/MyDrive/Colab-data/RDI-project/camel_data_toknized.csv",index_col=0)
```

```python
df["entities"]=df["entities"].apply(lambda x :literal_eval(x))
df["sentence"]=df["sentence"].apply(lambda x :literal_eval(x))
```

```python
sample_line = df.sentence
true_entities = df.entities
#sentence = simple_word_tokenize(sample_line)
pred = ner.predict(sample_line)
```

```python
true = list(true_entities)
```

```python
for i,data in enumerate(pred):
  print(pred[i],"--", true_entities[i])
```

```python
from nervaluate import Evaluator
```

```python
evaluator = Evaluator(true, pred, tags=['LOC', 'PER',"MIS","ORG"], loader="list")

results, results_by_tag = evaluator.evaluate()
```
